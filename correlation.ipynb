```python
import pandas as pd
```

```python
ee = pd.read_pickle('Final\\enfactor_array.pickle')
```

```python
x = pd.read_pickle('Final\\X_df.pkl')
y= pd.read_pickle('Final\\y_cri_df.pkl')
```

```python
y_consq = pd.read_pickle('Final\\y_consequence_df.pkl')
```

```python
new = pd.DataFrame()
```

```python
new[['dist_euclid',	'own_speed',	'target_speed',	'own_course',	'target_course']] = x
```

```python
new[['cri']] = y
```

```python
new[['Enfactors']] = ee
```

```python
new[['ECRI-W']] = y_consq
```

```python
new
```

```python
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as pplt
```

```python
cc = new[['dist_euclid', 'own_speed', 'target_speed', 'own_course', 'target_course', 'Enfactors', 'ECRI-W']].corr()
```

```python
pplt.figure(figsize=(10, 8))
sns.heatmap(cc, annot=True, cmap='Blues', linewidths=0.01)
#fig.tight_layout()
pplt.savefig("Corr Matrix Processed.png", bbox_inches='tight', dpi=300)
pplt.show()
```
